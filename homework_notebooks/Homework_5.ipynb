```python
plt.scatter(data[0], data[1])
plt.xlabel('Нулевой столбец')
plt.ylabel('Первый столбец')
plt.show()
```

По графику видим 1 выброс, зададим новую выборку без него

```python
data_new = data.drop(np.where(data[0] > 80000)[0], inplace=False)
```

```python
plt.scatter(data_new[0], data_new[1])
plt.xlabel('Нулевой столбец')
plt.ylabel('Первый столбец')
plt.show()
```

```python
print('Выборка с выбросами')
kk, p_value = st.pearsonr(data[0], data[1])
print('Коэффициент корреляции = {}, p_value = {}'.format(kk, p_value))
```

Поскольку p_value > 0.05 гипотеза о независимости выборок не отклоняется, что означает что они линейно независимы


```python
print('Обработанная выборка')
kk, p_value = st.pearsonr(data_new[0], data_new[1])
print('Коэффициент корреляции = {}, p_value = {}'.format(kk, p_value))
```

Поскольку p_value < 0.05 гипотеза о независимости выборок отклоняется , что означает что они линейно зависимы
